# finding query pattern ,selectivity,cardinality for db

```python

import psycopg2

# Database connection details
host = 'localhost'
port = '5432'
database = 'amazonn'
user = 'postgres'
password = 'hari'

# Connect to the PostgreSQL database
conn = psycopg2.connect(host=host, port=port, database=database, user=user, password=password)

# Create a cursor object to execute queries
cur = conn.cursor()

# Execute the query to retrieve data from the "flipkart" table
cur.execute('SELECT * FROM flipkart')

# Fetch all the rows from the result set
rows = cur.fetchall()

# Close the cursor and connection
cur.close()
conn.close()

# Get column names from the cursor description
column_names = [desc[0] for desc in cur.description]

# Calculate cardinality, selectivity, and query pattern for each attribute
cardinality = {}
selectivity = {}
query_pattern = {}

# Initialize attribute counters
total_rows = len(rows)
attribute_count = len(rows[0])  # Assuming all rows have the same number of attributes

# Iterate over the columns
for i in range(attribute_count):
    attribute_values = set()  # Using a set to store unique values for cardinality calculation
    attribute_name = column_names[i]

    # Iterate over the rows and collect attribute values
    for row in rows:
        attribute_values.add(row[i])

    # Calculate cardinality, selectivity, and query pattern
    attribute_cardinality = len(attribute_values)
    attribute_selectivity = attribute_cardinality / total_rows
    attribute_query_pattern = attribute_count / attribute_cardinality

    # Store the results
    cardinality[attribute_name] = attribute_cardinality
    selectivity[attribute_name] = attribute_selectivity
    query_pattern[attribute_name] = attribute_query_pattern

# Print the results
for attribute_name in cardinality:
    print(f"Attribute: {attribute_name}")
    print(f"Cardinality: {cardinality[attribute_name]}")
    print(f"Selectivity: {selectivity[attribute_name]}")
    print(f"Query Pattern: {query_pattern[attribute_name]}")
    print()
#cArdinality=unique ones
#selectivity=card/total records
#query=attributes count/cardinality
```

# statistics of the tables like frequent value and cardinality

```python

import psycopg2
from tabulate import tabulate

def find_best_index_attribute(table_name, host, port, database, user, password):
    conn = psycopg2.connect(
        host=host,
        port=port,
        dbname=database,
        user=user,
        password=password
    )

    cur = conn.cursor()

    # Retrieve column information from the table
    cur.execute(f"SELECT column_name, data_type FROM information_schema.columns WHERE table_name = '{table_name}'")
    columns = cur.fetchall()

    statistics = []

    for column in columns:
        column_name = column[0]
        data_type = column[1]

        # Calculate selectivity
        cur.execute(f"SELECT COUNT(DISTINCT {column_name}) FROM {table_name}")
        selectivity = cur.fetchone()[0]

        # Calculate cardinality
        cur.execute(f"SELECT COUNT(*) FROM {table_name}")
        total_rows = cur.fetchone()[0]
        cardinality = selectivity / total_rows

        # Identify the most frequent values
        cur.execute(f"SELECT {column_name}, COUNT(*) AS freq FROM {table_name} GROUP BY {column_name} ORDER BY freq DESC LIMIT 5")
        frequent_values = cur.fetchall()

        statistics.append((column_name, selectivity, cardinality, data_type, frequent_values))

    cur.close()
    conn.close()

    return statistics

def get_table_names(host, port, database, user, password):
    conn = psycopg2.connect(
        host=host,
        port=port,
        dbname=database,
        user=user,
        password=password
    )

    cur = conn.cursor()

    # Retrieve table names from the information_schema catalog
    cur.execute("""
        SELECT table_name
        FROM information_schema.tables
        WHERE table_schema = 'public'
        ORDER BY table_name
    """)

    table_names = cur.fetchall()

    cur.close()
    conn.close()

    return table_names

# Database connection details
host = 'localhost'
port = '5432'
database = 'amazonn'
user = 'postgres'
password = 'hari'

# Call the function to retrieve table names
table_names = get_table_names(host, port, database, user, password)

# Iterate over the table names and perform analysis for each table
for table in table_names:
    table_name = table[0]

    # Call the function to find the best index attribute for the current table
    statistics = find_best_index_attribute(table_name, host, port, database, user, password)

    # Prepare the data for tabular display
    data = []
    for stat in statistics:
        column_name = stat[0]
        selectivity = stat[1]
        cardinality = stat[2]
        data_type = stat[3]
        frequent_values = ', '.join([str(value[0]) for value in stat[4]])

        data.append([column_name, selectivity, cardinality, data_type, frequent_values])

    # Print the statistics for the current table in tabular format
    print(f"Table: {table_name}")
    headers = ["Attribute", "Selectivity", "Cardinality", "Data Type", "Frequent Values"]
    print(tabulate(data, headers=headers, tablefmt="grid"))
    print()
```

# best index attribute for overall table based on stats(selectivity and cardinality)

```python


import psycopg2
import matplotlib.pyplot as plt

def find_best_index_attribute(table_name, conn):
    cur = conn.cursor()

    # Retrieve column information from the table
    cur.execute(f"SELECT column_name, data_type FROM information_schema.columns WHERE table_name = '{table_name}'")
    columns = cur.fetchall()

    statistics = []

    for column in columns:
        column_name = column[0]
        data_type = column[1]

        # Exclude the primary key column
        if column_name == 'product_id':
            continue

        # Calculate selectivity
        cur.execute(f"SELECT COUNT(DISTINCT {column_name}) FROM {table_name}")
        selectivity = cur.fetchone()[0]

        # Calculate cardinality
        cur.execute(f"SELECT COUNT(*) FROM {table_name}")
        total_rows = cur.fetchone()[0]
        cardinality = selectivity / total_rows

        statistics.append((column_name, selectivity, cardinality, data_type))

    cur.close()

    # Sort statistics by cardinality in descending order
    statistics.sort(key=lambda x: x[2], reverse=True)

    # Select the attribute with the highest cardinality
    if statistics:
        best_attribute = statistics[0]
        print(f"Table: {table_name}")
        print(f"Best Index Attribute: {best_attribute[0]}")
        print(f"Selectivity: {best_attribute[1]}")
        print(f"Cardinality: {best_attribute[2]}")
        print(f"Data Type: {best_attribute[3]}")
        print("-------------------------")
        return best_attribute[0]

    return None

def get_table_names(conn):
    cur = conn.cursor()

    # Retrieve table names from the information_schema catalog
    cur.execute("""
        SELECT table_name
        FROM information_schema.tables
        WHERE table_schema = 'public'
        ORDER BY table_name
    """)

    table_names = cur.fetchall()

    cur.close()

    return table_names

# Database connection details
host = 'localhost'
port = '5432'
database = 'amazonn'
user = 'postgres'
password = 'hari'

# Create a PostgreSQL connection
conn = psycopg2.connect(
    host=host,
    port=port,
    dbname=database,
    user=user,
    password=password
)

# Call the function to retrieve table names
table_names = get_table_names(conn)

# Lists to store the table names and corresponding index attributes
table_index_pairs = []

# Iterate over the table names and choose the index attribute for each table
for table in table_names:
    table_name = table[0]

    # Call the function to find the best index attribute for the current table
    index_attribute = find_best_index_attribute(table_name, conn)

    if index_attribute:
        table_index_pairs.append((table_name, index_attribute))

# Extract the table names and index attributes into separate lists
table_names = [pair[0] for pair in table_index_pairs]
index_attributes = [pair[1] for pair in table_index_pairs]



# Close the connection
conn.close()
```

# execution time for each attribute without indexing

```python



import psycopg2
import time
from decimal import Decimal
import matplotlib.pyplot as plt

# Establish a connection to the database
conn = psycopg2.connect(
    host="localhost",
    port="5432",
    user="postgres",
    password="hari",
    database="amazonn"
)

# Create a cursor object to execute SQL queries
cursor = conn.cursor()

# Define the attributes to test
attributes = ['product_id', 'product_name', 'description', 'price', 'category', 'brand', 'rating']

# Variables to store the execution times and attribute names
execution_times = []
attribute_names = []

# Perform queries and measure execution time
for attribute in attributes:
    total_time = Decimal('0.0')
    for _ in range(100):
        start_time = time.time()
        cursor.execute(f"SELECT {attribute} FROM amazon")
        total_time += Decimal(str(time.time() - start_time))
    execution_time = total_time / Decimal('100')  # Average execution time
    
    # Append the execution time and attribute name to the lists
    execution_times.append(execution_time)
    attribute_names.append(attribute)

# Close the cursor and connection
cursor.close()
conn.close()

# Create a bar chart
plt.bar(attribute_names, execution_times)
plt.xlabel('Attribute')
plt.ylabel('Execution Time (seconds)')
plt.title('Execution Time for Each Attribute')
plt.xticks(rotation=45)
plt.show()
```

```python
print("############################################  ##########################")
print("############################################ AUTO INDEXING ###################################################")
print("############################################# ##############################################")






import psycopg2
import re
from psycopg2 import sql
from datetime import datetime

def analyzeQuery(queries):
    conn = psycopg2.connect(
        host="localhost",
        port="5432",
        user="postgres",
        password="hari",
        database="amazonn"
    )
    cur = conn.cursor()

    results = []

    try:
        for query in queries:
            # Execute the query
            cur.execute(query)

            # Get the column names from the result metadata
            columns = [desc[0] for desc in cur.description]

            # Get the table name from the query using pattern matching
            table_match = re.search(r"FROM\s+([^\s;]+)", query, re.IGNORECASE)
            table = table_match.group(1) if table_match else None

            # Store the table and columns in a dictionary
            result = {"table": table, "columns": columns}
            results.append(result)

    except psycopg2.Error as e:
        print("Error executing the query:", str(e))

    finally:
        cur.close()
        conn.close()

    return results


def createIndexes(query_results):
    conn = psycopg2.connect(
        host="localhost",
        port="5432",
        user="postgres",
        password="hari",
        database="amazonn"
    )
    cur = conn.cursor()

    for result in query_results:
        table = result["table"]
        columns = result["columns"]

        if table is not None:
            for col in columns:
                col_name = col
                index_name = f"{table}_{col_name}_index"

                # Check if the index already exists
                cur.execute(f"SELECT indexname FROM pg_indexes WHERE tablename = '{table}' AND indexname = '{index_name}'")
                if cur.fetchone():
                    print(f"Index '{index_name}' on table '{table}' for column '{col_name}' already exists.")
                else:
                    create_index_query = f'CREATE INDEX "{index_name}" ON "{table}" ("{col_name}")'
                    cur.execute(create_index_query)
                    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
                    print(f"[{timestamp}] Created index '{index_name}' on table '{table}' for column '{col_name}'")
        else:
            print("Table name not found for query:", query)

    conn.commit()
    cur.close()
    conn.close()

# Example usage
queries = [
    "SELECT rating,product_name FROM amazon WHERE rating > 4.5;",
    "SELECT product_name, price FROM amazon WHERE price < 100;",
    "SELECT * FROM products WHERE category = 'Electronics'"
]

query_results = analyzeQuery(queries)

for result in query_results:
    print("Table:", result["table"])
    print("Columns:", result["columns"])
    print()

createIndexes(query_results)
```

```python
import psycopg2
from datetime import datetime

def analyzeQuery(query):
    conn = psycopg2.connect(
        host="localhost",
        port="5432",
        user="postgres",
        password="hari",
        database="amazonn"
    )
    cur = conn.cursor()

    # Execute the EXPLAIN ANALYZE command to get the query plan
    cur.execute("EXPLAIN (FORMAT JSON) " + query)
    rows = cur.fetchall()
    #print(len(rows))
    tables = {}
    columns = []

    # Extract table and column information from the query plan
    for row in rows:
        print(row)
        plan_json = row[0][0]['Plan']
        #print(plan_json)
        table = plan_json.get('Relation Name')
        #print(table)
        if table not in tables:
            tables[table] = []
           
        output = plan_json.get('Output', [])
    
        if isinstance(output, list):
            columns.extend(output)
            tables[table].extend(output)

    cur.close()
    conn.close()

    return {"tables": tables, "columns": columns}


def createIndexes(query_plans):
    conn = psycopg2.connect(
        host="localhost",
        port="5432",
        user="postgres",
        password="hari",
        database="amazonn"
    )
    cur = conn.cursor()

    for plan in query_plans:
        tables = plan["tables"]
        columns = plan["columns"]

        for table, cols in tables.items():
            for col in cols:
                col_name = col
                index_name = f"{table}_{col_name}_index"
                
                # Check if the index already exists
                cur.execute(f"SELECT indexname FROM pg_indexes WHERE tablename = '{table}' AND indexname = '{index_name}'")
                if cur.fetchone():
                    print(f"Index '{index_name}' on table '{table}' for column '{col_name}' already exists.")
                else:
                    create_index_query = f'CREATE INDEX "{index_name}" ON "{table}" ("{col_name}")'
                    cur.execute(create_index_query)
                    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
                    print(f"[{timestamp}] Created index '{index_name}' on table '{table}' for column '{col_name}'")

    conn.commit()
    cur.close()
    conn.close()


# Example usage
queries = [
    "SELECT * FROM amazon WHERE rating > 4.5;",
    "SELECT product_name, price FROM amazon WHERE price < 100;",
    "UPDATE amazon SET brand = 'Brand X' WHERE category = 'Category 1';"
]

query_plans = [analyzeQuery(query) for query in queries]

createIndexes(query_plans)
```

```python
import psycopg2
import re

def extractColumnsFromQuery(query):
    # Regular expression pattern to match column names
    pattern = r"(?i)\bset\s+(.*?)\s+where\b"
    
    # Extract column names using regex pattern
    match = re.search(pattern, query)
    if match:
        columns_string = match.group(1)
        columns = [col.strip().split('=')[0] for col in columns_string.split(",")]
        return columns
    else:
        return []

def analyzeQuery(queries):
    conn = psycopg2.connect(
        host="localhost",
        port="5432",
        user="postgres",
        password="hari",
        database="amazonn"
    )
    cur = conn.cursor()

    try:
        for query in queries:
            cur.execute(query)
            
            # Check if the query produces a result set
            if cur.description is not None:
                # Get the column names from the result metadata
                columns = [desc[0] for desc in cur.description]
                
                # Print the query and its columns
                print("Query:", query)
                print("Columns:", columns)
                print()
            else:
                # Extract column names from the query
                columns = extractColumnsFromQuery(query)
                
                # Print the query and its columns
                print("Query:", query)
                print("Columns:", columns)
                print()

    except psycopg2.Error as e:
        print("Error executing the query:", str(e))

    finally:
        cur.close()
        conn.close()

# Example usage
queries = [
    "SELECT * FROM amazon WHERE rating > 4.5;",
    "SELECT product_name, price FROM amazon WHERE price < 100;",
    "UPDATE amazon SET brand = 'Brand X' WHERE category = 'Category 1';"
]

analyzeQuery(queries)
```
